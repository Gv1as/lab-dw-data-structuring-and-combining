```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df4 = pd.read_csv(url)
df4.head()
```

```python
df4.columns
```

```python
# It is not clear if the monthly premium needs to be multiplied by the number of policies so i did not do that.
# Agents seem to rake in the highest anount of revenue.

display(df4.pivot_table(index="sales_channel", values="monthly_premium_auto", aggfunc="sum").round(2))
```

```python
# Both genders are in every education level quite close in CLV, with the biggest difference in high school or bellow. High school or bellow gives the highest CLV for males and for females a close seccond.
# Bachelor, college and docter seems to be grouped close toghether regarding clv.

df4.pivot_table(index="gender", columns="education", values="customer_lifetime_value", aggfunc="mean")
```
